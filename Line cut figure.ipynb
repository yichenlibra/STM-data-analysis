# Nanonispy Method

```python
linecut_list = []
for i in range(33,35):
    linecut_temp = nap.read.Grid('.\STM-1 Run #84 NbSe2 film\Grid Spectroscopy0'+str(i)+'.3ds')
    linecut_list.append(linecut_temp)
```

```python
bias_vals=linecut_list[0].signals.get("sweep_signal")
bias = bias_vals *1000
#print(bias)
```

```python
linecut_1=linecut_list[0].signals.get("Lockin X (V)")
print(len(linecut_1))
print(linecut_1.shape)
```

```python
plt.figure(figsize=(6,10))
for i in range(len(linecut_1)):
    plt.plot(bias, linecut_1[i][0]+0.5*i)
plt.xlabel("Bias (mV)",size=20)
plt.ylabel("dI/dV (a.u.)",size=20)
plt.title('Line cut(Grid 33)',size=25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.savefig(".\STM-1 Run #84 NbSe2 film\Linecut 33.png", dpi=300)
plt.close()
```
